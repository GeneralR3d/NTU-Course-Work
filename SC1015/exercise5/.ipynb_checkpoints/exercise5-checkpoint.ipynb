```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import plot_tree
sb.set()
```

### Problem 1 : Predicting CentralAir using SalePrice
In this problem, you will build a Classification Tree model to predict CentralAir using SalePrice and judge its accuracy.


a) Create appropriate datasets for Train and Test in an 80:20 ratio and fit two Classification Tree models (of max 
depth 2 and max depth 4) on the Train set to predict CentralAir using SalePrice. Print the tree in each case.


```python
houses=pd.read_csv("train.csv")
houses.head()
```

*see swarm plot for SalePrice vs CentralAir*

```python
f=plt.figure(figsize=(20,10))
sb.swarmplot(data=houses,x="SalePrice",y="CentralAir")
plt.show()
```

*extract out the predictor and response and split them using train_test_split()*

```python
SalePrice=pd.DataFrame(houses["SalePrice"])
CentralAir=pd.DataFrame(houses["CentralAir"])

X_train,X_test,y_train,y_test=train_test_split(SalePrice,CentralAir,test_size=0.2)

dectree2=DecisionTreeClassifier(max_depth=2)
dectree4=DecisionTreeClassifier(max_depth=4)

dectree2.fit(X_train,y_train)
dectree4.fit(X_train,y_train)

f,axes=plt.subplots(2,1,figsize=(30,30))
plot_tree(dectree2,filled=True,rounded=True,feature_names=["SalePrice"],class_names=['N','Y'],ax=axes[0])
plot_tree(dectree4,filled=True,rounded=True,feature_names=["SalePrice"],class_names=['N','Y'],ax=axes[1])
```

b) Print the confusion matrix on Train set and Test set for both the aforesaid models. Compute and print for both 
the trees the Classification Accuracy, True Positive Rate, False Positive Rate. Which of the two models is better?

*the maxdepth=2 tree first*

```python
y_train_pred = dectree2.predict(X_train)
y_test_pred = dectree2.predict(X_test)

matrix_train=confusion_matrix(y_train, y_train_pred)
matrix_test=confusion_matrix(y_test, y_test_pred)

# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model of depth=2 \tTrain Dataset")
print("Classification Accuracy \t:", dectree2.score(X_train, y_train))
print(f"True Positive Rate\t:{matrix_train[1,1]/matrix_train[1].sum()}")
print(f"False Positive Rate\t:{matrix_train[0,1]/matrix_train[0].sum()}")
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model of depth=2 \tTest Dataset")
print("Classification Accuracy \t:", dectree2.score(X_test, y_test))
print(f"True Positive Rate\t:{matrix_test[1,1]/matrix_test[1].sum()}")
print(f"False Positive Rate\t:{matrix_test[0,1]/matrix_test[0].sum()}")
print()

# Plot the Confusion Matrix for Train and Test
f, axes = plt.subplots(1, 2, figsize=(12, 4))
sb.heatmap(matrix_train,
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(matrix_test, 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])
axes[0].set_title("Train")
axes[1].set_title("Test")
plt.show()
```

*then for the maxdepth=4 tree*

```python
y_train_pred = dectree4.predict(X_train)
y_test_pred = dectree4.predict(X_test)

matrix_train=confusion_matrix(y_train, y_train_pred)
matrix_test=confusion_matrix(y_test, y_test_pred)

# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model of depth=4 \tTrain Dataset")
print("Classification Accuracy \t:", dectree4.score(X_train, y_train))
print(f"True Positive Rate\t:{matrix_train[1,1]/matrix_train[1].sum()}")
print(f"False Positive Rate\t:{matrix_train[0,1]/matrix_train[0].sum()}")
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model of depth=4 \tTest Dataset")
print("Classification Accuracy \t:", dectree4.score(X_test, y_test))
print(f"True Positive Rate\t:{matrix_test[1,1]/matrix_test[1].sum()}")
print(f"False Positive Rate\t:{matrix_test[0,1]/matrix_test[0].sum()}")
print()

# Plot the Confusion Matrix for Train and Test
f, axes = plt.subplots(1, 2, figsize=(12, 4))
sb.heatmap(matrix_train,
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(matrix_test, 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])
axes[0].set_title("Train")
axes[1].set_title("Test")
plt.show()
```

*Looking at the classification accuracy for both models, both are very similar and has not much difference, for both train and test data. The same is true for the true positive rate. However since the false positive rate for the model with depth=4 is slightly lower than the model with depth=2, for both train and test data, the model with depth=4 is slighty better*

### Problem 2 : Predicting CentralAir using Other Variables
Following the steps from the previous problem, build two new uni-variate Classification Tree models (of max depth 4) to 
predict CentralAir using the variables OverallQual and YearBuilt. Justify which of the variables is the best predictor

*use the  variable OverallQual first*

```python
OverallQual=pd.DataFrame(houses["OverallQual"])
CentralAir=pd.DataFrame(houses["CentralAir"])

X_train,X_test,y_train,y_test=train_test_split(OverallQual,CentralAir,test_size=0.2)


dectree=DecisionTreeClassifier(max_depth=4)
dectree.fit(X_train,y_train)

f=plt.figure(figsize=(30,30))
plot_tree(dectree,filled=True,rounded=True,feature_names=["OverallQual"],class_names=['N','Y'])
```

*now do YearBuilt*

```python
YearBuilt=pd.DataFrame(houses["YearBuilt"])
CentralAir=pd.DataFrame(houses["CentralAir"])

X_train,X_test,y_train,y_test=train_test_split(YearBuilt,CentralAir,test_size=0.2)


dectree=DecisionTreeClassifier(max_depth=4)
dectree.fit(X_train,y_train)

f=plt.figure(figsize=(15,6))
plot_tree(dectree,filled=True,rounded=True,feature_names=["YearBuilt"],class_names=['N','Y'])
```

```python
y_train_pred = dectree.predict(X_train)
y_test_pred = dectree.predict(X_test)

matrix_train=confusion_matrix(y_train, y_train_pred)
matrix_test=confusion_matrix(y_test, y_test_pred)

# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of YearBuilt Model \tTrain Dataset")
print("Classification Accuracy \t:", dectree.score(X_train, y_train))
print(f"True Positive Rate\t:{matrix_train[1,1]/matrix_train[1].sum()}")
print(f"False Positive Rate\t:{matrix_train[0,1]/matrix_train[0].sum()}")
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of YearBuilt Model \tTest Dataset")
print("Classification Accuracy \t:", dectree.score(X_test, y_test))
print(f"True Positive Rate\t:{matrix_test[1,1]/matrix_test[1].sum()}")
print(f"False Positive Rate\t:{matrix_test[0,1]/matrix_test[0].sum()}")
print()

# Plot the Confusion Matrix for Train and Test
f, axes = plt.subplots(1, 2, figsize=(12, 4))
sb.heatmap(matrix_train,
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(matrix_test, 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])
axes[0].set_title("Train")
axes[1].set_title("Test")
plt.show()
```

*Both models have very similar classification accuracy, for both train and test data. However since we are concerned with the error rate, which is the false positive rate in this case, the OverallQual Model has false positive rate which is significantly lower than the YearBuilt model, for both train and test data. Therefore we conclude that OverallQual Model is the better model and the variable OverallQual is a better predictor of CentralAir*

### Problem 3 : Understanding the Misclassified Samples
In this problem, you will consider finer details of the tree model and try to isolate the misclassified samples for a model.

(a) Print the Classification Tree of max depth 4 that you fit in Problem 1(a). Note that each leaf node of your tree 
represents a specific partition of your training data, based on certain conditions given by the splits in the tree.
Find out which leaf node of this tree has the highest number of False Positives (FP) in the training dataset and 
print all samples in the Train set (rows from the dataframe) that end up in this leaf node (partition) during fit

```python
SalePrice=pd.DataFrame(houses["SalePrice"])
CentralAir=pd.DataFrame(houses["CentralAir"])

X_train,X_test,y_train,y_test=train_test_split(SalePrice,CentralAir,test_size=0.2)


dectree=DecisionTreeClassifier(max_depth=4)
dectree.fit(X_train,y_train)

f=plt.figure(figsize=(20,6))
plot_tree(dectree,filled=True,rounded=True,feature_names=["SalePrice"],class_names=['N','Y'])
plt.show()
```

*By observing all leaf nodes with classification as 'Y' according to the printed tree, leaf node number 3 (from left to right) has 70 samples, with overall classification of 'Y', but has 25 values which the model predicts as 'Y', which is the highest number of false positives in a leaf node.*

*since the values in this leaf node has SalePrice<= 98150.0 and SalePrice >62750, we can just filter out based on these conditions*

```python
jointDF = pd.concat([X_train, y_train], axis = 1).reindex(X_train.index)
leafnode = jointDF[(jointDF['SalePrice']>62750) & (jointDF['SalePrice']<=98150.0)]
leafnode
```

*we get exactly 70 rows which are the 70 samples in that leaf node*

(b) Print ONLY the False Positive cases (rows from the dataframe) from the samples (rows) extracted in part (a).

```python
leafnodeFP=leafnode[leafnode['CentralAir']=='N']
leafnodeFP
```

```python
leafnodeFP['CentralAir'].value_counts()
```

*has 25 entries which correspond to the 25 false positives in the leaf node in the decision tree*
