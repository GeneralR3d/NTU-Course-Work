a)Which of these variables has the maximum number of outliers as per box-plot? How many outliers does it have?

```python
f,axes=plt.subplots(4,1,figsize=(20,30))
sb.boxplot(data=house1["LotArea"],ax=axes[0],orient='h')
sb.boxplot(data=house1["GrLivArea"],ax=axes[1],orient='h')
sb.boxplot(data=house1["TotalBsmtSF"],ax=axes[2],orient='h')
sb.boxplot(data=house1["GarageArea"],ax=axes[3],orient='h')
```

*From the points that lie outside of the box and whiskers, LotArea has the most outliers. To count the number of outliers, a function to find out number of data points that lie outside of 1.5X Interquartile range is needed.*

```python
def countOutlier(dataframe):
    Q1=dataframe.quantile(.25)
    Q3=dataframe.quantile(.75)
    InterQ=Q3-Q1
    outliers=dataframe[((dataframe<(Q1-1.5*InterQ)) | (dataframe>(Q3+1.5*InterQ)))]
    return outliers.count()[dataframe.columns[0]] #count() returns dataframe, then dataframe.columns[0] gives the name of the first column, which we can use as a index

LotArea=pd.DataFrame(house1["LotArea"])
GrLivArea=pd.DataFrame(house1["GrLivArea"])
TotalBsmtSF=pd.DataFrame(house1["TotalBsmtSF"])
GarageArea=pd.DataFrame(house1["GarageArea"])

print(f"LotArea has {countOutlier(LotArea)} outliers.")
print(f"GrLivArea has {countOutlier(GrLivArea)} outliers.")
print(f"TotalBsmtSF has {countOutlier(TotalBsmtSF)} outliers.")
print(f"GarageArea has {countOutlier(GarageArea)} outliers.")
```

*therefore LotArea has most outliers*


b)Which of these variables is the most skewed from a regular normal distribution? Is the skew positive or negative?

```python
f,axes=plt.subplots(4,1,figsize=(30,20))
sb.histplot(data=LotArea,ax=axes[0],kde=True)
sb.histplot(data=GrLivArea,ax=axes[1],kde=True)
sb.histplot(data=TotalBsmtSF,ax=axes[2],kde=True)
sb.histplot(data=GarageArea,ax=axes[3],kde=True)
```

*From the KDE, it is not obvious which variable has the most left or right skew or deviation from normal distribution, so calculating mean and median of each is needed.*

```python
def findmean(dataframe):
    return dataframe.mean()[dataframe.columns[0]]

def findmedian(dataframe):
    return dataframe.median()[dataframe.columns[0]]


print(f"LotArea mean: {findmean(LotArea)}, median: {findmedian(LotArea)}, difference: {findmean(LotArea)-findmedian(LotArea)}")
print(f"GrLivArea mean: {findmean(GrLivArea)}, median: {findmedian(GrLivArea)}, difference: {findmean(GrLivArea)-findmedian(GrLivArea)}")
print(f"TotalBsmtSF mean: {findmean(TotalBsmtSF)}, median: {findmedian(TotalBsmtSF)}, difference: {findmean(TotalBsmtSF)-findmedian(TotalBsmtSF)}")
print(f"GarageArea mean: {findmean(GarageArea)}, median: {findmedian(GarageArea)}, difference: {findmean(GarageArea)-findmedian(GarageArea)}")
```

*Since LotArea has the biggest difference between mean and median, it is most skewed from a normal distribution, since for a normal distribution,mean==median. In particular, LotArea has the greatest right skew.*

Problem 3 : Interesting Questions for EDA

b) Does the SalePrice of a house get affected by how recently it got Remodeled? Justify your answer using EDA..

'YearBuilt' and 'YearRemodAdd' related to this particular problem. However, since some houses are not remodeled ever since they were built, a new variable 'Age' can be defined



```python
# exclude houses that was never remodeled
df = houses.drop(
    houses[houses["YearBuilt"] == houses["YearRemodAdd"]].index
)
# exclude houses that was remodeled after it was sold
df = df.drop(df[df["YearRemodAdd"] > df["YrSold"]].index)

# calculate houseAge
Age = df["YrSold"] - df["YearRemodAdd"]
Age.head()
```

```python
f = plt.figure(figsize=(10, 5))
x, y = pd.Series(Age, name="Age"), pd.Series(df['SalePrice'], name="SalePrice")
sb.regplot(x = x, y = y)
```

From the above graph and the function output of findCorrelation we can see that the houseAge and SalePrice indeed has a negative correlation.

a)Does the SalePrice of a house get affected by whether it has a Garage or not? Justify your answer using EDA.

*Take out all the columns that are related to having a garage*

```python
house3=houses[['GarageType','GarageYrBlt','GarageFinish','GarageCars','GarageArea','GarageQual','GarageCond','SalePrice']]
house3.info()
```

*If we look for all the signs which a house has no Garage, ie GarageType,GarageType,GarageFinish,GarageQual or GarageCond is empty, OR when GarageCars =0 which means the garage has space for 0 cars, OR when GarageArea=0 which means the garage has 0 area, all these are signs of a missing garage and we will try to filter all of them out*

```python
mask=(house3['GarageType'].isnull()) | (house3['GarageYrBlt'].isnull()) | (house3['GarageFinish'].isnull()) | (house3['GarageCars']==0) | (house3['GarageArea']==0) | (house3['GarageQual'].isnull()) | (house3['GarageQual'].isnull())
mask
```

```python
house3.insert(loc=0,column="hasGarage",value=~mask,allow_duplicates=True)#the inverted version of the mask
house3
```

*Now that we have added the column for whether each house has a garage, we can then use catplot to find out if those with a garage has any signifcantly different saleprice based on the quartiles given by a boxplot*

```python
sb.catplot(data=house3,y='SalePrice',x='hasGarage',kind='box')
```

*From the boxplots it is clear that houses with garage has a much higher median, 25th quartile and 75th quartile saleprice than those without a garage, thus on average, those houses with garage will have a higher saleprice*
