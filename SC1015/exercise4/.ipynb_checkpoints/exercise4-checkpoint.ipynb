```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split #helps split data evenly and randomly
from sklearn.metrics import mean_squared_error #built in function to help perform Goodness-Of-Fit test
sb.set()
```

### Problem 1 : Predicting SalePrice using GrLivArea
Note : We observed during EDA that GrLivArea and SalePrice have a strong linear relationship with correlation 0.71.
In this problem, you will build a Linear Regression model to predict SalePrice using GrLivArea and judge its accuracy.

a) Create appropriate datasets for Train and Test in an 80:20 ratio and fit a Linear Regression model on the Train 
set to predict SalePrice using GrLivArea. Print the coefficients of your model and plot the regression line.

*read data and extract SalePrice and GrLivingArea*

```python
houses=pd.read_csv("train.csv")
houses.head()
```

```python
SalePrice=pd.DataFrame(houses['SalePrice'])
SalePrice.head()
```

```python
GrLivArea=pd.DataFrame(houses['GrLivArea'])
GrLivArea.head()
```

*split the data into 80-20*

```python
x_train, x_test, y_train, y_test = train_test_split(GrLivArea, SalePrice, test_size = 0.2)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

*start linear regression*

```python
linreg=LinearRegression()
linreg.fit(x_train,y_train)

print('Intercept of Regression \t: b = ', linreg.intercept_)
print('Coefficients of Regression \t: a = ', linreg.coef_)
print()
```

```python
y_train_prediction=linreg.predict(x_train)

f = plt.figure(figsize=(16, 8))
plt.xlabel("GrLivArea")
plt.ylabel("SalePrice")
plt.scatter(x_train, y_train)
plt.scatter(x_train, y_train_prediction, color = "r")
plt.show()
```

b) Check the Goodness of Fit of the model on the Train set and Prediction Accuracy of the model on the Test set.
Print the metrics for Goodness of Fit and Prediction Accuracy that you think are appropriate in each scenario.

*goodness of fit refers to how well the model fits the real data, thus the metric for measuring that would be the R^2, which is the ratio between ESS and TSS, ie how much of the variance in the SalePrice can be predicted by the GrLivArea*

```python
# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model GrLivArea \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(x_train, y_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_train, y_train_prediction))
print()

# Check the Goodness of Fit (on Test Data)
y_test_prediction=linreg.predict(x_test)
print("Goodness of Fit of Model GrLivArea \tTest Dataset")
GrLivArea_test_R2=linreg.score(x_test, y_test)
print("Explained Variance (R^2) \t:", GrLivArea_test_R2)
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_test, y_test_prediction))
print()
```

### Problem 2 : Predicting SalePrice using Other Variables
Following the steps from the previous problem, build two new uni-variate Linear Regression models to predict SalePrice
using the variables TotalBsmtSF and GarageArea, individually. Justify which of the three models is the best in this case

```python
GarageArea=pd.DataFrame(houses['GarageArea'])
x_train, x_test, y_train, y_test = train_test_split(GarageArea, SalePrice, test_size = 0.2)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
linreg=LinearRegression()
linreg.fit(x_train,y_train)

print('Intercept of Regression \t: b = ', linreg.intercept_)
print('Coefficients of Regression \t: a = ', linreg.coef_)
print()

y_train_prediction=linreg.predict(x_train)

f = plt.figure(figsize=(16, 8))
plt.xlabel("GarageArea")
plt.ylabel("SalePrice")
plt.scatter(x_train, y_train)
plt.scatter(x_train, y_train_prediction, color = "r")
plt.show()

# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model for GarageArea \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(x_train, y_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_train, y_train_prediction))
print()

# Check the Goodness of Fit (on Test Data)
y_test_prediction=linreg.predict(x_test)
print("Goodness of Fit of Model for GarageArea \tTest Dataset")
GarageArea_test_R2=linreg.score(x_test, y_test)
print("Explained Variance (R^2) \t:", GarageArea_test_R2)
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_test, y_test_prediction))
print()
```

*comparing all R^2 values for all 3 variables on test data*





```python
print("Explained Variance (R^2) GrLivArea Test Dataset \t:", GrLivArea_test_R2)
print("Explained Variance (R^2) TotalBsmtSF Test Dataset \t:", TotalBsmtSF_test_R2)
print("Explained Variance (R^2) GarageArea Test Dataset \t:", GarageArea_test_R2)
```

*Since GrLivArea has the highest R^2 value, the model using GrLivArea as predictor has the best Goodness of Fit to the actual test data, i.e. GrLivArea explains most of the variation in SalePrice, among the 3, thus GrLivArea is the best model for predicting SalePrice*

### Problem 3 : Refining the Models to Predict SalePrice
In this problem, you will consider finer details of the dataset and the variables to refine the model to predict SalePrice

a) Find the houses (rows) that are â€œoutliersâ€ for GrLivArea and/or SalePrice. This means outliers for GrLivArea
UNION outliers for SalePrice in a set notation. Remove all these â€œoutliersâ€ from the dataset so that it is clean

```python
house3=houses[["GrLivArea","SalePrice"]]
house3.info()
```

```python
#removing GrLivArea outliers
Q1=house3["GrLivArea"].quantile(.25)
Q3=house3["GrLivArea"].quantile(.75)
InterQ=Q3-Q1
house3_filtered=house3[((house3["GrLivArea"]>(Q1-1.5*InterQ)) & (house3["GrLivArea"]<(Q3+1.5*InterQ)))]
house3_filtered.info()
```

```python
#removing SalePrice outliers
Q1=house3["SalePrice"].quantile(.25)
Q3=house3["SalePrice"].quantile(.75)
InterQ=Q3-Q1
house3_filtered=house3[((house3["SalePrice"]>(Q1-1.5*InterQ)) & (house3["SalePrice"]<(Q3+1.5*InterQ)))]
house3_filtered.info()
```

(b) In the outlier-free dataset, create Train and Test sets with an 80:20 ratio, and fit a Linear Regression model on 
the Train set to predict SalePrice using GrLivArea. Print the model coefficients and plot the regression line.

```python
GrLivArea_fil=pd.DataFrame(house3_filtered['GrLivArea'])
SalePrice_fil=pd.DataFrame(house3_filtered['SalePrice'])
x_train, x_test, y_train, y_test = train_test_split(GrLivArea_fil, SalePrice_fil, test_size = 0.2)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
linreg=LinearRegression()
linreg.fit(x_train,y_train)

print('Intercept of Regression \t: b = ', linreg.intercept_)
print('Coefficients of Regression \t: a = ', linreg.coef_)
print()

y_train_prediction=linreg.predict(x_train)

f = plt.figure(figsize=(16, 8))
plt.xlabel("GarageArea without outliers")
plt.ylabel("SalePrice without outliers")
plt.scatter(x_train, y_train)
plt.scatter(x_train, y_train_prediction, color = "r")
plt.show()
```

(c) Check the Goodness of Fit of the model on the Train set and Prediction Accuracy of the model on the Test set.
Do you think this model is better than the model obtained in Problem 1 for the same variables? Briefly justify.

```python
# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model GrLivArea cleaned \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(x_train, y_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_train, y_train_prediction))
print()

# Check the Goodness of Fit (on Test Data)
y_test_prediction=linreg.predict(x_test)
print("Goodness of Fit of Model GrLivArea cleaned \tTest Dataset")
print("Explained Variance (R^2) \t:", linreg.score(x_test, y_test))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_test, y_test_prediction))
print()
```

*Compared to the values for the "uncleaned" GrLivArea model from **Problem 1**, the Explained Variance (R^2) for both train and test data are quite similar, however, the Mean Squared Error (MSE) of the "cleaned" GrLivArea model is much lower and better, thus this model with outliers removed is a slightly better model for predicting SalePrice.*
